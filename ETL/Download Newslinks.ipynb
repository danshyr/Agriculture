```python
# -*- coding: utf-8 -*- 
import requests
from bs4 import BeautifulSoup
from math import ceil
import re

oil_list = open('oil.txt','r')                           #開啟檔案
for line in oil_list.readlines():
    pagelink = line.strip()
    m = re.match(r".+\?id=.+&.*",pagelink)              #用正規表示法篩選網頁
    if m is not None:    
        rear_substring = pagelink.split("?")[1]           #切出所要的ID作為檔案名稱
        param = rear_substring.split("&")[0]        
        request_get = requests.get(pagelink)
        
        response = request_get.text.encode('utf8')   
        
        soup = BeautifulSoup(response)
        title = soup.select('#tdTitle')[0]                        #取得標題
        article = soup.select('#tdContent')[0]                   #取得內文
        #print  title                                               #測是列印
        oil_detail = open("chinaoil/%s.txt"%(param),'w')         #將不同網頁存成成文字檔
        oil_detail.write(title.prettify('utf8')+"\n"+article.prettify('utf8'))
        oil_detail.close()
oil_list.close() #關閉檔案
    
    
```
