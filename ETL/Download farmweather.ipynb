```python
# -*- coding: utf-8 -*- 

import requests
from bs4 import BeautifulSoup

weather_detail = open('weatherKS.txt','w')  #開啟空檔案
                                       #因為中央氣象局農業觀測資料網頁是用年月日在字尾，所以用迴圈跑網頁出來
for i in range(1999,2016):             #年
    for j in range(1,13):              #月
        if j <10:                      #用判斷是在小於10的月份補上0，網頁才能讀取
            str_month = "0"+ str(j)            
        else:
            str_month =  str(j)            
        for k in range(1,4):          #上旬中旬下旬為1.2.3
            link = "http://www.cwb.gov.tw/V7/climate/agri/Data/10day/72Q01_"+str(i)+"_"+str_month+"_"+str(k)+".htm"
            print 
            res = requests.get(link)                    #因為編碼問題所以要編UTF8兩次
            res.encoding = 'utf8'
            response = res.text.encode('utf8')
            soup = BeautifulSoup(response)                    

            column = soup.select('.tab01')             #取得column名子並解寫入檔案
            for title in column:                             
                #print title.text                      #測試列印
                weather_detail.write(title.text.encode("utf8")+",")
            weather_detail.write(",")
            number = soup.select('.num')
            for num in number:                         #取得數據並解寫入檔案
                #print num.text                         #測試列印
                weather_detail.write(num.text.encode("utf8")+",")
            weather_detail.write("\n")
weather_detail.close()                            #結束
```
