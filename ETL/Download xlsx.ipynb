```python
# -*- coding: utf-8 -*- 
import requests
from bs4 import BeautifulSoup
import shutil                          #載檔案用的套件

res = requests.get("http://amis.tari.gov.tw/MapManager/PageContent.aspx?type=C&id=27")
res.encoding ="utf8"
response_text = res.text.encode('utf8') 
soup = BeautifulSoup(response_text)

links = soup.select('a')[4:6]               #因為格式問題所以要分2個迴圈去抓
for link in links:
    href = link['href']                       #取超聯結內容
    filename =href[-10:]                      #取超聯結的尾巴找出檔名
    #print filename    
    res = requests.get(href, stream=True)
    f = open("weatherxls/%s"%(filename), 'wb')    #依序寫入檔案
    res.raw.decode_content = True                 #網路上查到的用法
    shutil.copyfileobj(res.raw, f)                #網路上查到的用法
    f.close()   

links = soup.select('a')[6:-43]
for link in links:
    href = "http://amis.tari.gov.tw/MapManager/"+link['href']    
    #print href
    filename =href[-10:]
    res = requests.get(href, stream=True)
    f = open("weatherxls/%s"%(filename), 'wb')
    res.raw.decode_content = True
    shutil.copyfileobj(res.raw, f)     
    f.close()   
```
