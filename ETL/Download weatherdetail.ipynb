```python
# -*- coding: utf-8 -*- 
import requests
from bs4 import BeautifulSoup

for i in range(2012,2016):             
    for j in range(1,13):              
        if j <10:                      
            str_month = "0"+ str(j)            
        else:
            str_month =  str(j)       
        for k in range(1,31):
            if k <10:                      
                str_day = "0"+ str(k)            
            else:
                str_day =  str(k)              
            link = "http://e-service.cwb.gov.tw/HistoryDataQuery/DayDataController.do?command=viewMain&station=467440&datepicker="+str(i)+"-"+str_month+"-"+str_day
            #print link            
            res = requests.get(link)
            res.encoding ="utf8"
            response_text = res.text.encode('utf8')  
            
            soup = BeautifulSoup(response_text)
            filename = str(i)+str_month+str_day
            weather_history = open("weatherKS/%s.txt"%(filename),'w')     
            
            header = soup.select('tr td')[:3]
            for head in header:
                weather_history.write(head.text.encode('utf8')+",")
                #print head.text
            weather_history.write("\n")
    
            first_row = soup.select('.second_tr')
            for title in first_row:
                title_content = title.select('th')
                for real_content in title_content:
                    weather_history.write(real_content.text.encode('utf8')+",")
                    #print real_content.text          
            weather_history.write("\n")
            second_rows = soup.select('tr')[3:]
            for rows in second_rows:
                rowscontent =rows.select('td')
                weather_history.write("\n")
                for content in rowscontent:
                    weather_history.write(content.text.encode('utf8')+",")               
                    print content.text
weather_history.close()
```
